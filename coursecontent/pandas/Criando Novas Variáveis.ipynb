## Criando Novas Variáveis

```
import pandas as pd
```

```
data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Curso Pandas/datasets/aluguel_residencial_filtrado.csv', sep=';')
```

```
data.head(10)
```

```
data['Valor Bruto'] = data['Valor'] + data['Condominio']+ data['IPTU']
```

```
data.head(10)
```

```
data['Valor m2'] = data['Valor']/data['Area']
```

```
data.head(10)
```

```
data['Valor m2'] = data['Valor m2'].round(2)
```

```
data.head(10)
```

```
data['Valor m2 Bruto'] = (data['Valor Bruto']/data['Area']).round(2)
```

```
data.head()
```

```
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```
data['Tipo Agregado'] = data['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```
data.head()
```

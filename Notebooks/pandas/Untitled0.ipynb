```
import pandas as pd
```

```
data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Curso Pandas/datasets/aluguel_residencial_filtrado.csv', sep=';')
```

```
data.head(5)
```

```
# 1 e 2
# 3 e 4
# 5 e 6
# 7 ou mais
classes = [0,2,4,6,100]
```

```
quartos = pd.cut(data.Quartos,classes)
```

```
quartos
```

```
pd.value_counts(quartos)
```

```
labels = ['1 e 2 quartos','3 e 4 quartos','5 e 6 quartos','7 quartos ou mais']
```

```
quartos = pd.cut(data.Quartos,classes,labels=labels)
```

```
pd.value_counts(quartos)
```

```
quartos = pd.cut(data.Quartos,classes,labels=labels, include_lowest = True)
```

```
pd.value_counts(quartos)
```
